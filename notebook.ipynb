```python
import micrograd.nn as nn
import numpy as np
from micrograd.engine import Tensor
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

```python
x = Tensor(X)
```

```python
class model(nn.Module):
    def __init__(self):
        self.softmax = nn.softmax()

    def forward(self, x):
        return self.softmax(x)
```

```python
from graphviz import Digraph
```

```python
def trace_text(root):
    visited = set()

    def build(v, depth=0):
        if v not in visited:
            visited.add(v)

            print("  "*depth, v._op, "shape:", v.data.shape)

            for child in v._prev:
                build(child, depth+1)

    build(root)
```

```python
test = model()
```

```python
y = test(x)
```

```python
y.backward()
```

```python
trace_text(y)
```
